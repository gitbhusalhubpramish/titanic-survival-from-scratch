```python
df = pd.read_csv("/kaggle/input/datasets/ashishkumarjayswal/titanic-datasets/titanic.csv")
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Fare"] = df["Fare"].fillna(df["Fare"].median())
df["Sex"] = df["Sex"].map({"male":0, "female":1})
```

```python
x = df[["Pclass", "Age", "Sex", "Fare"]]
y = df["Survived"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
w = np.array([np.random.rand()*0.001]*len(x.columns))
b = np.random.rand()
w,b
```

```python
import math
```

```python
def sigmodi(z):
    return 1/(1+math.e**(-z))
```

```python
def reg(x):
    return x@w+b
```

```python
def predict(x):
    return sigmodi(reg(x))
predict(xtrain)
```

```python
def loss(x,y):
    p = predict(x)
    p = np.clip(p, 1e-15, 1-1e-15)
    return -np.mean(y*np.log(p) + (y-1)*np.log(1-p))
```

```python
def train():
    global w,b
    prevloss = loss(xtrain,ytrain)
    for epo in range(10000):
        p = predict(xtrain)
        # p = np.clip(p, 1e-15, 1-1e-15)
        dw = (xtrain.T @ (p-ytrain)) / len(xtrain)
        db = np.mean((p-ytrain))
        learning_rate = 0.01
        w-= learning_rate*dw
        b-= learning_rate*db  
        l = loss(xtrain, ytrain)
        if abs(prevloss-l) <= 1e-15:
            print(prevloss,l, dw,db)
            print(epo)
            break
        prevloss = l
```

```python
train()
loss(xtrain,ytrain)
```

```python
loss(xtest,ytest)
```

```python
from sklearn.metrics import accuracy_score
pred = predict(xtest)
pred = (pred >= 0.5).astype(int)
accuracy_score(pred,ytest)
# pred
```
